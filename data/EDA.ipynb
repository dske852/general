```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
%matplotlib inline
#import warnings
import os
import re
#warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('data.csv')
```

```python
path = "./eda_data"
files = [file for file in os.listdir(path) if not file.startswith('.')] # Ignore hidden files

all_months_data = pd.DataFrame()

for file in files:
    current_data = pd.read_csv(path+"/"+file)
    all_months_data = pd.concat([all_months_data, current_data])
    
all_months_data.to_csv("all_data_copy.csv", index=False)
```

```python
df=all_months_data.copy()
```

```python
df.drop_duplicates(inplace=True, ignore_index=True)
df.head()
```

```python
#numerical & categorical columns
numeric_features = [feature for feature in df.columns if df[feature].dtype != 'O']
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']
```

```python
print(numeric_features)
```

```python
for col in categorical_features:
    print(col," : ",df[col].unique())
```

```python
df.loc[:,'Quantity Ordered'] = df.loc[:,'Quantity Ordered'].apply(lambda x: re.sub("[^0123456789\.]","", x) )
df.head()
```

```python
df.loc[:,'Price Each'] = df.loc[:,'Price Each'].apply(lambda x: re.sub("[^0123456789\.]","", x) )
df.head()
```

```python
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
df['Price Each'] = pd.to_numeric(df['Price Each'])
```

```python
df.info()
```

```python
df.loc[:,'Order Date'] = df.loc[:,'Order Date'].apply(lambda x: np.nan if x == "Order Date" else x )
```

```python
df.loc[:,'Product'] = df.loc[:,'Product'].apply(lambda x: np.nan if x == "Product" else x )
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df['Date'] = df['Order Date'].dt.month
```

```python
df['Year'] = df['Order Date'].dt.year
```

```python
df['Day'] = df['Order Date'].dt.day
```

```python
df['Time'] = df['Order Date'].dt.hour
```

```python
df.info()
```

```python
df.head()
```

```python
df['City']=df['Purchase Address'].str.split(",", expand=True)[1]
```

```python
df['State']=df['Purchase Address'].str.split(",", expand=True)[2].str.split(" ", expand=True)[1]
```

```python
product_mapping ={'USB-C Charging Cable':'Cable', 'Bose SoundSport Headphones':'Headphones', 'Google Phone':'Phone',
 'Wired Headphones':'Headphones', 'Macbook Pro Laptop':'Laptop', 'Lightning Charging Cable':'Cable', '27in 4K Gaming Monitor':'Monitor',
 'AA Batteries (4-pack)':'Batteries','Apple Airpods Headphones':'Headphones', 'AAA Batteries (4-pack)':'Batteries', 'iPhone':'Phone', 'Flatscreen TV':'Tv/Monitor',
  '27in FHD Monitor':'Tv/Monitor', '20in Monitor': 'Tv/Monitor', 'LG Dryer':'Home gadget', 'ThinkPad Laptop':'Laptop', 'Vareebadd Phone':'Phone','LG Washing Machine':'Home gadget',
   '34in Ultrawide Monitor': 'Tv/Monitor'
 }

df['Product_merged'] = df.Product.map(product_mapping)
```

```python
df['Sales'] = df['Quantity Ordered'] * df['Price Each']
```
